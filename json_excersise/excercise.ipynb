```python
import pandas as pd


df = pd.read_csv('../data/json_exercise.csv')

df
```

```python
import json
```

```python
def load_json(value):
    try:
        return json.loads(value)
    except (json.JSONDecodeError, TypeError):
        return None

# Aplica la función a la columna 'Value' y convierte los resultados en un array
data = df['Value'].apply(load_json).array

# Imprime los datos cargados como objetos JSON
print(data)
```

```python
# Creamos un DataFrame a partir del array de numpy
df = pd.DataFrame(data, columns=['json_data'])

# Creamos un DataFrame 'atts' con las claves 'AttID', 'Cnt' y 'TS'
atts = df['json_data'].apply(pd.Series).dropna(subset=['AttID', 'Cnt', 'TS'])

# Creamos un DataFrame 'vars' con las claves 'VarID', 'Value', 'LRL', 'LSL', 'Target', 'URL' y 'USL'
vars = df['json_data'].apply(pd.Series).dropna(subset=['VarID', 'Value', 'LRL', 'LSL', 'Target', 'URL', 'USL'])




```

```python
# Función para cargar el JSON
def load_json(value):
    try:
        return json.loads(value)
    except (json.JSONDecodeError, TypeError):
        return None

# Cargamos el archivo CSV
df = pd.read_csv('../data/json_exercise.csv')

```

```python
df.head()
```

```python
gum = df[df['Value'].str.contains('Gum')]
gum
```

```python
gum_2 = gum[['Value', 'TS', 'Tag']].copy()
gum_2
```

```python
nombres = gum_2['Tag'].unique()
print(nombres)
```

```python
unique_values = gum_2['Value'].unique()
print(unique_values)
```

```python
# Extraemos el valor después de los dos puntos en cada celda de la columna 'Value'
gum_2['Value'] = gum_2['Value'].str.split(':').str[-1].str.strip()
gum_2['Value'] = gum_2['Value'].str.replace('"', '').str.replace('{', '').str.replace('}', '')


# Añadimos las columnas 'TS' y 'Tag' a 'gum'
gum_2[['TS', 'Tag']] = df[['TS', 'Tag']]

# Creamos el DataFrame final 'gum_2' con las columnas 'Gum_Acacia', 'TS' y 'Tag'
gum_3 = gum_2[['Value', 'TS', 'Tag']].copy()

gum_3
```

```python
gum_3.shape
```

```python
# Eliminar caracteres no numéricos de la columna 'TS'
gum_3['TS'] = gum_3['TS'].replace(r'\D', '', regex=True)

# Convertir la columna 'TS' a tipo fecha utilizando el formato adecuado
gum_3['TS'] = pd.to_datetime(gum_3['TS'], format='%Y%m%d%H%M%S')
```

```python
gum_3
```

```python
# Cambiar el nombre de la columna 'Value' a 'Gum_Acacia'
gum_3.rename(columns={'Value': 'Gum_Acacia'}, inplace=True)

# Mostrar el DataFrame 'gum_3' actualizado
gum_3
```
